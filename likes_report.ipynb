The data in this report is made by Instagram based on the user's activity. The data was created to build a portfolio on the Instagram user that tracks what kind of media they consume and like. This data can be used to note patterns and behaviors about the user and can futher add to the data on genreal Instagram users of similar demographics. 

This data is reliable in the sense that it provides the account that posted the content and a link to the post. It is important that the post in it's original state when liked by the user is documented. Unfortunately, the posts can be deleted so the reference to it can be broken. Also, it might help to know how many other people liked the post before the user, just to get a sense of how popular a particular post is. 

```python
import json
import pandas as pd
import matplotlib as plt
```

```python
path = "/Users/biancaamoako/Data_EMAT/Likes Report/liked_posts.json"

with open(path) as j:
    likes_dict = json.load(j)

likes_dict.keys()
```

```python
likes_df = pd.DataFrame(likes_dict["likes_media_likes"])
likes_df.head()
```

```python
#[x[0]["timestamp"] for x in likes_df["string_list_data"]]
```

```python
likes_df["Time of Like"] = [x[0]["timestamp"] for x in likes_df["string_list_data"]]
likes_df.head()
```

```python
likes_df.groupby("title").count()
```

```python
likes_df.groupby("title").count().sort_values("string_list_data", ascending = False).head(10)\
.plot(y = "Time of Like", kind = "bar", color = "firebrick")
```

### Hypothesis about Personal TikTok Data

Initial Hypothesis: I share TikToks through the app the most compared to other avenues of sharing. By counting each method of sharing, I will be able to see the most common avenues of sharing I use. The attribute "Method" withing the Shared History section has a few different values that it can be.

```python
path = "/Users/biancaamoako/Data_EMAT/user_data_tiktok.json"

with open(path) as j:
    share_dict = json.load(j)

share_dict.keys()
```

```python
share_dict["Activity"]["Share History"]["ShareHistoryList"]
```

```python
shares_df = pd.DataFrame(share_dict["Activity"]["Share History"]["ShareHistoryList"])
shares_df.head(10)
```

The data lists the videos that I have shared since March 20, 2024. The type of content shared is listed, all of them being videos, the link is included, and the menthod of sharing is listed. Each shared video(observation) has it's own row and each attribute of an observation has it's own column, making this pretty tidy data. The only downside is that the method "more" is vague. 

```python
shares_df.groupby("Method").count().sort_values("Date", ascending = False)
```

```python
shares_df.groupby("Method").count().sort_values("Date", ascending = False).plot(y = "Date", kind = "bar", color = "firebrick")
```
